```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from weather_app.configs import TRANSFORMED_DATA_DIR

FORECASTS_PATH = TRANSFORMED_DATA_DIR / "forecasts.parquet"
```

```python
df = pd.read_parquet(FORECASTS_PATH)
df
```

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.lineplot(data=df, x="time", y="temperature_2m", hue="city", ax=ax)
ax.set_title("Temperaturas: Rio de Janeiro x Juiz de Fora")
plt.show()
```
